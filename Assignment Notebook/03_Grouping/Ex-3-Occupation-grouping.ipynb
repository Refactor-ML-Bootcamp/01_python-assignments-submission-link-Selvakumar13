```python
#1. Import the necessary libraries
import pandas as pd
```

```python
#2. Import the dataset from this address.
#3. Assign it to a variable called users.
users=pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',sep='|',index_col='user_id')
```

```python
users
```

```python
#4. Discover what is the mean age per occupation
mean_age=users.groupby('occupation')['age'].mean()
mean_age
```

```python
#5. Discover the Male ratio per occupation and sort it from the most to the least

male_ratio=users.groupby("occupation")['gender'].value_counts()
male_ratio
```

```python
#6. For each occupation, calculate the minimum and maximum ages
min_max_age=users.groupby('occupation')['age'].describe()[['min','max']]
min_max_age
```

```python
#7. For each combination of occupation and gender, calculate the mean age
mean_age=users.groupby(["occupation","gender"])['age'].mean()
mean_age
```

```python
#8. For each occupation present the percentage of women and men
male_ratio=users.groupby('occupation')['gender'].value_counts().div(users.gender.count())
male_ratio
```
